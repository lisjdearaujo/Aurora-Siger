## 5.1 Organização e Descrição da Telemetria

Para a validação do pré-lançamento da nave Aurora Siger, foram definidos cinco
parâmetros de telemetria, cada um com faixas seguras baseadas em dados de
missões espaciais reais.

---

### Temperatura Interna (temp_interna)
Tipo: (real/float)
Representa a temperatura da cabine e dos sistemas eletrônicos embarcados.  
**Faixa segura:** 18°C a 27°C  
Valores fora dessa faixa indicam risco térmico aos equipamentos e à tripulação,
resultando em abortagem imediata.


---

### Temperatura Externa (temp_externa)
Tipo: (real/float)
Corresponde à temperatura do casco externo da nave.  
**Faixa segura:** -50°C a +60°C  
Temperaturas além desses limites comprometem a integridade dos materiais
estruturais e dos painéis térmicos.

---

### Integridade Estrutural (integridade)
Tipo: (bool)
Variável booleana que representa o estado físico da nave.  
**1 = sem falhas detectadas | 0 = comprometimento detectado**  
Qualquer falha nesse parâmetro aborta automaticamente a missão.

---

### Nível de Energia (energia)
Tipo: float
Indica o percentual de carga das baterias principais.  
**Faixa segura:** mínimo 95%  
Abaixo desse valor, o risco de falha de sistemas críticos durante o lançamento
é inaceitável.

---

### Pressão dos Tanques (pressao)
Tipo: float
Monitora a pressão do combustível nos tanques propulsores.  
**Faixa segura:** 300 a 350 kPa  
Pressão fora dessa faixa pode causar falha na ignição ou ruptura dos tanques.

---

### Status dos Módulos Críticos (modulos)
Tipo: categórico discreto (ou conjunto de booleanos)
Indica se todos os módulos essenciais estão operacionais.  
**1 = todos os sistemas online | 0 = falha em ao menos um módulo**  
Qualquer módulo offline impede imediatamente a decolagem.

## 5.2 Algoritmo de Verificação

O algoritmo foi construído seguindo uma estrutura sequencial de decisões.
Cada parâmetro é verificado em ordem — se qualquer um estiver fora da faixa
segura, a decolagem é abortada imediatamente. Somente se todos os parâmetros
estiverem dentro dos limites, o sistema autoriza o lançamento.

O fluxograma (disponível no link do drive) foi desenvolvido com as seguintes verificações, nessa ordem:

1. Temperatura interna entre 18°C e 27°C
2. Temperatura externa entre -50°C e 60°C
3. Integridade estrutural igual a 1
4. Nível de energia maior ou igual a 95%
5. Pressão dos tanques entre 300 e 350 kPa
6. Módulos críticos todos online (valor 1)

Link do Fluxograma: https://drive.google.com/file/d/1eSqc4i8fLPn1I4BtSuSj8NXe4EzFIOHT/view?usp=sharing

## A seguir, o pseudocódigo correspondente ao fluxograma:

```
ALGORITMO VerificacaoDecolagem

VARIÁVEIS
    temp_interna : real
    temp_externa : real
    integridade  : inteiro
    energia      : real
    pressao      : real
    modulos      : inteiro

INÍCIO

    // Verificação 1: temperatura interna
    ESCREVA "Informe a temperatura interna"
    LEIA temp_interna
    SE (temp_interna < 18) OU (temp_interna > 27) ENTÃO
        ESCREVA "Decolagem abortada!"
        FIM_ALGORITMO
    SENÃO
        ESCREVA "Temperatura interna dentro do permitido."
    FIM_SE

    // Verificação 2: temperatura externa
    ESCREVA "Informe a temperatura externa"
    LEIA temp_externa
    SE (temp_externa < -50) OU (temp_externa > 60) ENTÃO
        ESCREVA "Decolagem abortada!"
        FIM_ALGORITMO
    SENÃO
        ESCREVA "Temperatura externa dentro do permitido."
    FIM_SE

    // Verificação 3: integridade estrutural
    ESCREVA "Integridade estrutural (1=OK, 0=FALHA)"
    LEIA integridade
    SE (integridade <> 1) ENTÃO
        ESCREVA "Decolagem abortada!"
        FIM_ALGORITMO
    SENÃO
        ESCREVA "Integridade estrutural OK."
    FIM_SE

    // Verificação 4: nível de energia
    ESCREVA "Informe o nível de energia (%)"
    LEIA energia
    SE (energia < 95) ENTÃO
        ESCREVA "Decolagem abortada!"
        FIM_ALGORITMO
    SENÃO
        ESCREVA "Energia suficiente."
    FIM_SE

    // Verificação 5: pressão dos tanques
    ESCREVA "Informe a pressão dos tanques"
    LEIA pressao
    SE (pressao < 300) OU (pressao > 350) ENTÃO
        ESCREVA "Decolagem abortada!"
        FIM_ALGORITMO
    SENÃO
        ESCREVA "Pressão dentro do permitido."
    FIM_SE

    // Verificação 6: módulos críticos
    ESCREVA "Módulos críticos OK? (1=SIM, 0=NÃO)"
    LEIA modulos
    SE (modulos <> 1) ENTÃO
        ESCREVA "Decolagem abortada!"
        FIM_ALGORITMO
    SENÃO
        ESCREVA "Módulos OK."
    FIM_SE

    // Todas as verificações passaram
    ESCREVA "PRONTO PARA DECOLAR!"

FIM
```

## 5.4 Análise Energética

Com base nos dados reais de consumo de naves espaciais, foi calculada a
autonomia energética da nave Aurora Siger para a fase de decolagem.

| Parâmetro | Valor |
|---|---|
| Capacidade total das baterias | 500 kWh |
| Carga atual | 97% |
| Energia disponível | 485 kWh |
| Consumo estimado na decolagem | 300 kWh |
| Perdas energéticas (estimada em 5%) | 15 kWh |
| Energia restante após decolagem | 170 kWh |

**Cálculo:**
- Energia disponível = 500 × 0,97 = **485 kWh**
- Energia consumida + perdas = 300 + 15 = **315 kWh**
- Saldo após decolagem = 485 - 315 = **170 kWh**

**Conclusão:** A nave possui energia suficiente para a decolagem com margem
de segurança de 170 kWh, equivalente a aproximadamente 14% da capacidade total.

Essas escolhas têm consequências diretas quando há vidas envolvidas. Um valor de corte mal definido, uma verificação ignorada ou uma ordem de prioridade errada no algoritmo podem custar uma missão inteira.
